```python
# write a function to check if the given input number is prime or not if prinme return true or else return false in python
```

```python
def is_prime(n):
   
    if n >= 2:
        for i in range(2, n):
            if not (n % i):
                return False
        return True  
    else:
        return False
```

```python
# alternate  process
def is_prime(num):
    for i in range(2, num//2 + 1, 1):
        if num % i == 0:
            return False
    return True
    
```

```python
is_prime(4)
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
n=int(input("Input a number to compute the factiorial : "))
print(factorial(n))
```

```python
# write a python function that take a list and returns a new list with unique elements of the first list
```

```python

def f(l):

    x = []

    for a in l:

        if a not in x:

            x.append(a)

    return x

print(f([1,2,3,3,3,3,4,5]))
```

```python
def count_upper_lower(string):
    upper = 0
    lower = 0
    for letter in string:
        if letter.isupper():
            upper = upper + 1
        elif letter ==" ":
            continue
        if letter.islower():
            lower = lower + 1
    return upper, lower
            
```
