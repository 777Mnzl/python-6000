# Logical operators

* and : all expression must be True 
* or : atleast one must be True
* not : none should be True 



|A    |B     |
|---: |---:  |
|True |False |
|False|True  |

|A|B|A and B|
|---:|---:|---:|
|True|True|True|
|True|False|False|
|False|True|False|
|False|False|False|


|A|B|A or B|
|---:|---:|---:|
|True|True|True|
|True|False|True|
|False|True|True|
|False|False|False|


```python
num = int(input("enter any number:"))
if num > 0 and num <10:
    print("both condition are true")
else:
    print("condition are false")
    
```

```python
num = int(input("enter any number:"))
if num > 0 or num <10:
    print("one condition are true")
else:
    print("condition are false")
```
